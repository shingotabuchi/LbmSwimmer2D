```python
def occilation_pos(t,omega0,gamma,initVel,initPos):
    if omega0 > gamma:
        omega = np.sqrt(omega0**2 - gamma**2)
        return np.exp(-gamma*t)*(initPos * np.cos(omega*t) + ((gamma*initPos + initVel)/omega)*np.sin(omega*t))
    if omega0 == gamma:
        return ((initVel + gamma * initPos)*t + initPos) * np.exp(-gamma*t)
    if omega0 < gamma:
        gammaPlus = gamma + np.sqrt(gamma**2-omega0**2)
        gammaMinus = gamma - np.sqrt(gamma**2-omega0**2)
        C1 = (-initVel - gammaMinus*initPos)/(gammaPlus-gammaMinus)
        C2 = (initVel + gammaPlus*initPos)/(gammaPlus-gammaMinus)
        return C1*np.exp(-gammaPlus*t) + C2*np.exp(-gammaMinus*t)
```

```python
def occilation_vel(t,omega0,gamma,initVel,initPos):
    if omega0 > gamma:
        omega = np.sqrt(omega0**2 - gamma**2)
        return -gamma*occilation_pos(t,omega0,gamma,initVel,initPos) + np.exp(-gamma*t)*(-initPos * omega*np.sin(omega*t) + (gamma*initPos + initVel)*np.cos(omega*t))
    if omega0 == gamma:
        return -gamma*occilation_pos(t,omega0,gamma,initVel,initPos) + (initVel + gamma * initPos) * np.exp(-gamma*t)
    if omega0 < gamma:
        gammaPlus = gamma + np.sqrt(gamma**2-omega0**2)
        gammaMinus = gamma - np.sqrt(gamma**2-omega0**2)
        C1 = (-initVel - gammaMinus*initPos)/(gammaPlus-gammaMinus)
        C2 = (initVel + gammaPlus*initPos)/(gammaPlus-gammaMinus)
        return -gammaPlus*C1*np.exp(-gammaPlus*t) -gammaMinus*C2*np.exp(-gammaMinus*t)
```

```python
def convergetime(dt,omega0,gamma,initVel,initPos):
    t = 0
    while True:
        pos = occilation_pos(t,omega0,gamma,initVel,initPos)
        vel = occilation_vel(t,omega0,gamma,initVel,initPos)
        if np.abs(pos) <= 0.1 and np.abs(vel) <= 0.1:
            return t
        t += dt
```
