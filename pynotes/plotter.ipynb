```python
import pandas as pd
import numpy as np
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('figure',  figsize=(8, 8))
```

```python
data_texts = []
data_names = []
data_arrays = []
rootdir = os.path.abspath(os.getcwd()) + "\\texts"
index = 0
bincount = 10
skipSet = set({0,1,2,3,4})
for file in os.listdir(rootdir):
    d = os.path.join(rootdir, file)
    data_names.append(file)
    data_texts.append(d)
    data_arrays.append(np.loadtxt(d))
    if index not in skipSet: 
        print(np.average(data_arrays[index]))
        print(np.min(data_arrays[index]))
        print(np.max(data_arrays[index]))
        plt.hist(data_arrays[index],bins = bincount,label = file)
    index+=1
plt.legend()
plt.show()
```

```python
data_theta_times = []
data_theta_arrays = []
rootdirtheta = os.path.abspath(os.getcwd()) + "\\initThetas"
index = 0
bincount = 10
for file in os.listdir(rootdirtheta):
    dTheta = os.path.join(rootdirtheta, file)
    d = os.path.join(rootdir, file)
    data_theta_times.append(np.loadtxt(d))
    data_theta_arrays.append(np.loadtxt(dTheta))

    plt.scatter(data_theta_arrays[index],data_theta_times[index],label = file)
    index+=1
plt.legend()
plt.show()
```
